```python
def main():
    # Prompt the user for temperature in Fahrenheit
    fahrenheit = input("Enter temperature in Fahrenheit: ")

    # Convert the input to a float
    fahrenheit = float(fahrenheit)

    # Convert to Celsius using the given formula
    celsius = (fahrenheit - 32) * 5.0 / 9.0

    # Print the result
    print("Temperature: " + str(fahrenheit) + "F = " + str(celsius) + "C")

# Required line to call the main function when script is run directly
if __name__ == '__main__':
    main()
```
