```python
import string
string.ascii_lowercase
import random
words = ['python', 'hangman', 'challenge']
word = random.choice(words)
guessed = set()
alphabet = set(string.ascii_lowercase)
def hangman():
    word = random.choice(words)
    word_letters = set(word)
    guessed_letters = set()
    lives = 6

    while len(word_letters) > 0 and lives > 0:
        print(f"You have {lives} lives left. Used letters: {' '.join(guessed_letters)}")

        word_list = [letter if letter in guessed_letters else '-' for letter in word]
        print('Current word:', ' '.join(word_list))

        guess = input("Guess a letter: ").lower()

        if guess in string.ascii_lowercase - guessed_letters:
            guessed_letters.add(guess)
            if guess in word_letters:
                word_letters.remove(guess)
            else:
                lives -= 1
                print("Incorrect guess.")
        elif guess in guessed_letters:
            print("You already guessed that letter.")
        else:
            print("Invalid input.")

    if lives == 0:
        print(f"Sorry, you lost. The word was {word}.")
    else:
        print(f"Congrats! You guessed the word: {word}!")

hangman()
```
