```python
def main():
    # Dictionary with fruits and their prices
    fruits = {
        'apple': 1.5,
        'durian': 50,
        'jackfruit': 80,
        'kiwi': 1,
        'rambutan': 1.5,
        'mango': 5
    }

    total_cost = 0.0

    # Ask how many of each fruit the user wants
    for fruit_name in fruits:
        price = fruits[fruit_name]
        amount = int(input(f"How many ({fruit_name}) do you want?: "))
        total_cost += amount * price

    # Print total formatted to 2 decimal places
    print(f"Your total is ${total_cost:.2f}")

# Required line to run the main function
if __name__ == '__main__':
    main()
```
