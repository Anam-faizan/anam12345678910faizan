```python
def main():
    numbers = [1, 2, 3, 4]  # Initial list of numbers

    # Loop through the list using indices and double each value
    for i in range(len(numbers)):
        numbers[i] = numbers[i] * 2

    print(numbers)  # Output: [2, 4, 6, 8]

# This provided line is required at the end of the Python file
if __name__ == '__main__':
    main()
```
