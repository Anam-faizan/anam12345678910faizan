```python
import streamlit as st

# Sidebar for navigation
st.sidebar.title("Navigation")
page = st.sidebar.radio("Go to", ["Home", "About", "Contact"])

# Home page
if page == "Home":
    st.title("üè† Welcome to My Website")
    st.write("This is a simple Python website built using Streamlit.")

    # Example: Small calculator
    st.header("üßÆ Add Two Numbers")
    a = st.number_input("Enter first number:")
    b = st.number_input("Enter second number:")
    if st.button("Calculate"):
        st.success(f"The result is {a + b}")

# About page
elif page == "About":
    st.title("‚ÑπÔ∏è About")
    st.write("This website was built with Streamlit. It's great for building Python web apps fast!")

# Contact page
elif page == "Contact":
    st.title("üì¨ Contact")
    st.write("Reach me at: [your-email@example.com](mailto:your-email@example.com)")
```
