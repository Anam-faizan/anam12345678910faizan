```python
"""
Simulate rolling two dice, and print the result of each roll and the total.
"""

import random

# Number of sides on a die
NUM_SIDES = 6

def main():
    # Optionally set a seed for consistent test results
    # random.seed(1)

    # Simulate rolling two dice
    die1 = random.randint(1, NUM_SIDES)
    die2 = random.randint(1, NUM_SIDES)
    total = die1 + die2

    # Display results
    print(f"Dice have {NUM_SIDES} sides each.")
    print(f"First die: {die1}")
    print(f"Second die: {die2}")
    print(f"Total of two dice: {total}")

# Required to call the main function when the script runs
if __name__ == '__main__':
    main()
```
