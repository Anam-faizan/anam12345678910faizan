```python
class Student:
    def __init__(self, name, marks):
        self.name = name      # Assigning name using self
        self.marks = marks    # Assigning marks using self

    def display(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)

# Example usage:
student1 = Student("Alice", 92)
student1.display()
```
