```python
MINIMUM_HEIGHT = 50  # Minimum height required to ride

def main():
    height_input = input("How tall are you? ")

    # Convert input to float and check against minimum height
    if height_input:
        height = float(height_input)
        if height >= MINIMUM_HEIGHT:
            print("You're tall enough to ride!")
        else:
            print("You're not tall enough to ride, but maybe next year!")

# This ensures the main function is called when the script is executed
if __name__ == '__main__':
    main()
```
