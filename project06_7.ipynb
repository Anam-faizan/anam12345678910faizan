```python
class Employee:
    def __init__(self, name, salary, ssn):
        self.name = name        # Public variable
        self._salary = salary   # Protected variable (by convention)
        self.__ssn = ssn        # Private variable (name mangled)

# Create an object
emp = Employee("Ali", 50000, "123-45-6789")

# Access public variable
print("Name:", emp.name)


# Access protected variable (allowed, but not recommended)
print("Salary (protected):", emp._salary)

# Try accessing private variable directly (will raise AttributeError)
try:
    print("SSN (private):", emp.__ssn)
except AttributeError as e:
    print("Error accessing __ssn directly:", e)

# Access private variable using name mangling
print("SSN (accessed with name mangling):", emp._Employee__ssn)
```
