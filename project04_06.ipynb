```python
import time

def countdown_timer():
    seconds = int(input("Enter the time in seconds: "))

    while seconds > 0:
        mins, secs = divmod(seconds, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        print(timer, end='\r')
        time.sleep(1)
        seconds -= 1

    print("Time's up!")

countdown_timer()
```
