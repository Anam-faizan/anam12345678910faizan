```python
class A:
    def show(self):
        print("Show from class A")

class B(A):
    def show(self):
        print("Show from class B")

class C(A):
    def show(self):
        print("Show from class C")

class D(B, C):  # D inherits from B and C
    pass

# Create object of class D and call show()
d = D()
d.show()

# Display the MRO
print("MRO:", [cls.__name__ for cls in D.__mro__])
```
