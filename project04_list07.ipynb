```python
def main():
    lst = []  # Create an empty list to hold user input

    val = input("Enter a value: ")  # Prompt the user for the first value
    while val != "":  # Keep looping until the user presses Enter without input
        lst.append(val)  # Add the value to the list
        val = input("Enter a value: ")  # Ask again for the next input

    print("Here's the list:", lst)  # Print the final list

# Required to start the program
if __name__ == '__main__':
    main()
```
