```python
ADULT_AGE = 18  # U.S. age where someone is legally considered an adult

def is_adult(age: int):
    if age >= ADULT_AGE:
        return True
    return False

########## No need to edit code beyond this point :) ##########

def main():
    age = int(input("How old is this person?: "))  # We need to convert the input to an integer
    print(is_adult(age))

if __name__ == "__main__":
    main()
```
