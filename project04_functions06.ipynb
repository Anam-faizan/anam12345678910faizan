```python
def main():
    for i in range(10, 20):
        if is_odd(i):
            print(f"{i} odd")
        else:
            print(f"{i} even")

def is_odd(value: int):
    """
    Checks to see if a value is odd. If it is, returns True.
    """
    return value % 2 == 1

# This provided line is required at the end of
# Python file to call the main() function.
if __name__ == '__main__':
    main()
```
