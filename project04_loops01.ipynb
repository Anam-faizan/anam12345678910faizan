```python
MAX_TERM_VALUE = 10000  # The maximum value for the Fibonacci sequence

def main():
    curr_term = 0  # The 0th Fibonacci Number
    next_term = 1  # The 1st Fibonacci Number

    # Print the Fibonacci sequence as long as the current term is less than or equal to MAX_TERM_VALUE
    while curr_term <= MAX_TERM_VALUE:
        print(curr_term)
        # Calculate the next term in the sequence
        term_after_next = curr_term + next_term
        # Update the current term and the next term
        curr_term = next_term
        next_term = term_after_next


# Python boilerplate
if __name__ == '__main__':
    main()
```
