```python
def get_user_info():
    # Asking the user for their first name
    first_name = input("What is your first name?: ")

    # Asking for their last name
    last_name = input("What is your last name?: ")

    # Asking for their email address
    email_address = input("What is your email address?: ")

    # Returning all the data as a tuple
    return first_name, last_name, email_address

def main():
    # Calling get_user_info to get the user data
    user_data = get_user_info()

    # Printing out the received user data (tuple)
    print("Received the following user data:", user_data)

if __name__ == "__main__":
    # Running the main function
    main()
```
