```python
"""
Program: dicesimulator
----------------------
Simulate rolling two dice, three times. Prints
the results of each die roll. Demonstrates how
variable scope works in Python.
"""

import random

# Number of sides on each die
NUM_SIDES = 6

def roll_dice():
    """
    Simulates rolling two dice and prints their total.
    This function uses local variables die1 and die2,
    which are separate from any die1 defined elsewhere.
    """
    die1 = random.randint(1, NUM_SIDES)
    die2 = random.randint(1, NUM_SIDES)
    total = die1 + die2
    print("Rolled dice: {} + {} = {}".format(die1, die2, total))

def main():
    die1 = 10  # Local to main()
    print("die1 in main() starts as:", die1)

    # Roll dice three times
    roll_dice()
    roll_dice()
    roll_dice()

    # die1 in main() remains unchanged
    print("die1 in main() is still:", die1)

# Entry point of the program
if __name__ == '__main__':
    main()
```
