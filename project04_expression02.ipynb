```python
# Constant: speed of light in meters per second
C = 299792458  # m/s

def main():
    # Prompt user for mass in kilograms
    mass_in_kg = float(input("Enter kilos of mass: "))

    # Calculate energy using Einstein's formula: E = m * c^2
    energy_in_joules = mass_in_kg * (C ** 2)

    # Display the results
    print("\ne = m * C^2...")
    print("m = " + str(mass_in_kg) + " kg")
    print("C = " + str(C) + " m/s")
    print(f"{energy_in_joules} joules of energy!")

# This provided line is required at the end of the Python file
# to call the main() function.
if __name__ == '__main__':
    main()
```
