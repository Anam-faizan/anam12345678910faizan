```python
"""
Program: add2numbers
--------------------
This program asks the user for two integers and prints their sum.
"""

def main():
    print("This program adds two numbers.")

    # Prompt and read first number
    num1 = input("Enter first number: ")
    num1 = int(num1)

    # Prompt and read second number
    num2 = input("Enter second number: ")
    num2 = int(num2)

    # Calculate the sum
    total = num1 + num2

    # Print the result
    print("The total is " + str(total) + ".")

# Required line to call main function when script is run directly
if __name__ == '__main__':
    main()
```
