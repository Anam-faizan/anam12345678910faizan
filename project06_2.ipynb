```python
class Counter:
    count = 0  # Class variable to keep track of object count

    def __init__(self):
        Counter.count += 1  # Increment count when a new object is created

    @classmethod
    def show_count(cls):
        print("Number of objects created:", cls.count)

# Example usage:
obj1 = Counter()
obj2 = Counter()
obj3 = Counter()

Counter.show_count()
```
